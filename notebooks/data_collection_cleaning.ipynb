## Set data folder structure
Copy folder to 'data/interim' 

```python
import shutil
import errno
import os

def copy(src, dest):
    try:
        shutil.copytree(src, dest)
    except OSError as e:
        # If the error was caused because the source wasn't a directory
        if e.errno == errno.ENOTDIR:
            shutil.copy(src, dest)
        else:
            print('Directory not copied. Error: %s' % e)
```

```python
for folder in os.listdir('../data/raw'):
    if not folder.startswith('download'):
        copy('../data/raw/{}'.format(folder), '../data/interim/{}'.format(folder))
```
